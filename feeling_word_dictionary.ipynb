# Time for some extra NLP

For funsies, we can take a look at how different these feeling adjectives are from one another in word2vec. Then compare those distances to the distances between a bag of words representation of their definitons, then even a tfidf of their definitions. This could illustrate the value-add of vector representations - and the success of either method would motivate the utility of these advanced NLP techniques. 

Subsequently, words could be assembled into an adjacency matrix:
    words with a sufficiently high similarity score would have edges between them
    words dissimilar would be part of different components (possibly?!)

Given an adjacency matrix of feels, would we get clusters of emotion words? Would there be components to the graph or would a path exist between any pair of feels? Stay tuned in this data adventure!

```python
import pandas as pd
feelSimilarityDict = {"feelPairs":[], "similarity":[]}
feelPairs = []
similarities = []
# Creating a dictionary where the keys are the pairs of feel words I 
# am comparing and the values are the smiliarity values of the words
for token1 in tokens:
    for token2 in tokens:
        # Ensuring that the words are not spaces, and that word vectors exist for them
        if token1 != token2 and str(token1) != " " and str(token2) != " " \
        and token1.has_vector and token2.has_vector:
            feelPairs.append(str(token1.text)+" "+str(token2.text)) 
            similarities.append(token1.similarity(token2))
            feelAdjacencyMtrx.loc[str(token1).strip(),str(token2).strip()]= token1.similarity(token2)

feelSimilarityDict["feelPairs"] = feelPairs
feelSimilarityDict["similarity"] = similarities
# Creating a dictionary to feed a data frame with the similarity data
feelSimilarityDf = pd.DataFrame(feelSimilarityDict)
```
